```python
import pandas as pd
import os.path
```

```python
data_directory = '/Users/albertosantos/Downloads/MaxQuant_Output/'
prot_directory = data_directory
data_files = ['proteinGroups.txt', 'Peptides.txt','Oxidation (M)Sites.txt']
analytical_samples = 'P0000003_analytical_samples.csv'
timepoints = {'pre':'-1','post':'0', '3m':'90', '1y':'365', '4y':'1460'}
```

```python
data = {}
for f in data_files:
    path = os.path.join(data_directory, prot_directory+f)
    data[f] = pd.read_csv(path, sep='\t',low_memory=False)
    
mapping = pd.read_csv(os.path.join('/Users/albertosantos/Development/Clinical_Proteomics_Department/ClinicalKnowledgeGraph(CKG)/data/imports/experiments/P0000003/',analytical_samples))
```

```python
mapping_dict = pd.Series(mapping.ID.values,index=mapping.external_id).to_dict()
```

```python
for dataset in data:
    print(dataset)
    new_cols = []
    for c in data[dataset].columns:
        new_col = c
        if "intensity " in c.lower() or "experiment " in c.lower():
            aux = c.split(' ')
            sample_timepoint = aux[-1].split('_')[-3:]
            if '_'.join(sample_timepoint[1:]) in mapping_dict and sample_timepoint[2] in timepoints:
                sample = mapping_dict['_'.join(sample_timepoint[1:])]
                timepoint = timepoints[sample_timepoint[2]]
                replicate = sample_timepoint[0]
                new_col = " ".join(aux[0:-1])+" "+replicate+"_"+sample+"_"+timepoint
        new_cols.append(new_col)         
    data[dataset].columns = new_cols
    data[dataset].to_csv("new_"+dataset, sep='\t', index=False, doublequote=False)
```
